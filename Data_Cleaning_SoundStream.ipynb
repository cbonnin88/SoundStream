```python
import polars as pl
import gdown as gd
```

```python
url = 'https://drive.google.com/uc?id=1w7f_WAqF4PSL1c7gwS2_eM1nuTYl2JgD'
gd.download(url,'user_behavior_raw.csv',quiet=True)
```

```python
df_soundstream_raw = pl.read_csv('user_behavior_raw.csv')
```

# **Phase 1: Data Cleaning**

```python
# 1. Converting signup_date to datetime objects
# 2. Ensuring categorical columns are handled correctly

df_soundstream = df_soundstream_raw.with_columns([
    pl.col('signup_date').str.to_date('%Y-%m-%d'),
    pl.col('subscription_status').cast(pl.Categorical),
    pl.col('primary_device').cast(pl.Categorical)
])
```

```python
# 3. Checking for any missing values
print('Missing Values per Column:')
display(df_soundstream.null_count())
```

```python
df_soundstream.write_csv('soundstream_user_behavior_claned.csv')
```
