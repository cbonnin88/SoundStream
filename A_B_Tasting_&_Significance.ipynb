```python
import polars as pl
from scipy import stats
import plotly.graph_objects as go
import gdown as gd
```

```python
url = 'https://drive.google.com/uc?id=1TkZ3_vwEH6zCZAa6QZWzhtFhKreHinbk'
gd.download(url,'soundstream_user_behavior_cleaned.csv',quiet=True)
```

```python
df_soundstream = pl.read_csv('soundstream_user_behavior_cleaned.csv')
```

# **A/B Testing**

```python
# Filter Groups

premium_ratings = df_soundstream.filter(pl.col('subscription_type') == 'Premium Individual')['music_suggestion_rating_1_to_5'].to_numpy()
free_ratings = df_soundstream.filter(pl.col('subscription_type')== 'Free')['music_suggestion_rating_1_to_5'].to_numpy()
```

```python
# Perform Independent T-Test
t_stat, p_value = stats.ttest_ind(premium_ratings,free_ratings,equal_var=False)
```

```python
# Visualizing Distribution Difference
fig_ab = go.Figure()
fig_ab.add_trace(go.Histogram(x=premium_ratings,name='Premium Individual',opacity=0.75))
fig_ab.add_trace(go.Histogram(x=free_ratings,name='Free Tier',opacity=0.75))

fig_ab.update_layout(barmode='overlay',title='A/BS Test: Suggestion Ratings Distribution')
fig_ab.show()
```

```python
print(f'T-Statistic: {t_stat:.4f}')
print(f'P-Value: {p_value:.4f}')
```

```python
if p_value < 0.05:
  print('Conclusion: The difference is STATISTICALLY SIGNIFICANT. Premium users have a different satisfaction profile')
else:
  print('Conclusion: The difference is NOT SIGNIFICANT. The rating gap may be due to random choice')
```
